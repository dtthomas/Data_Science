```python
z1 = [1,np.nan,np.nan,4,5,np.nan,7,np.nan] #nan means not a number
z2 = [1,1,1,1,2,2,2,2]
```

```python
z1
```

```python
z2
```

```python
df1 = pd.DataFrame({'Cricket':z1,
                   'Tennis':z2}) #dictionary
```

```python
df1
```

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df1.to_csv("Missing_value_eg.csv")  #Write the data frome in to a csv file
```

```python
df1.dropna()  #method 1: drop the missing value records. Not a recommended method.
```

```python
df1
```

```python
df1.fillna(0)  #Method 2: Replace missing values with one single value
```

```python
df1.fillna(df1['Cricket'].mean()) #Method 3: Replace missing value with the mean/mode value of column
```

```python
#Method 4: Find some relationship among different columns and choose the missing values based on the related column values present.
#eg: Define the missing values of Cricket column based on the value of Tennis column
#group the Cricket column values based on the values of Tennis column and find mean within groups.
```

```python
df1['Cricket'].fillna(df1.groupby(['Tennis'])['Cricket'].transform(np.mean))
```
