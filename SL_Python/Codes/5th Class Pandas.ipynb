Working with the missing value

```python
import numpy as np
dictionary = {'First Score':[100,90,np.nan,95],
             'Second Score':[30,45,56,np.nan],
             'Third Score':[np.nan,40,80,98]}

df = pd.DataFrame(dictionary)

print(df)
```

```python
print(df.isnull())
```

```python
df.fillna(df.mean())
```

```python
df.dropna()
```

Grouping Rows in pandas


```python
# importing Pandas 
import pandas as pd 

# example dataframe 
example = {'Team':['Arsenal', 'Manchester United', 'Arsenal', 
				'Arsenal', 'Chelsea', 'Manchester United', 
				'Manchester United', 'Chelsea', 'Chelsea', 'Chelsea'], 
					
		'Player':['Ozil', 'Pogba', 'Lucas', 'Aubameyang', 
					'Hazard', 'Mata', 'Lukaku', 'Morata', 
										'Giroud', 'Kante'], 
										
		'Goals':[5, 3, 6, 4, 9, 2, 0, 5, 2, 3] } 

df = pd.DataFrame(example) 

print(df)
```

```python
total_goals = df['Goals'].groupby(df['Team'])
print(total_goals.std())
```
